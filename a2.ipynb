```python
import pandas as pd

# Assuming df is your pandas DataFrame with numeric columns containing null values

# Identify numeric columns with null values
numeric_cols_with_null = demographic_df.select_dtypes(include=['number']).columns[demographic_df.isnull().any()]

# Calculate median for each numeric column with null values
median_values =demographic_df[numeric_cols_with_null].median()

# Replace null values with median for each numeric column
demographic_df[numeric_cols_with_null] = demographic_df[numeric_cols_with_null].fillna(median_values)

# Show updated DataFrame
demographic_df
```
