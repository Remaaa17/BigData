```python
import pandas as pd
# Specify the file path
file_path = r"C:\Users\abdel\OneDrive\Desktop\Sub\SystemAnalysis\SA.Project\Crude suicide rates.edit.csv"

# Load the CSV file into a Pandas DataFrame
Crude_suicide_df = pd.read_csv(file_path)

# Display the first few rows of the DataFrame
print("Crude_suicide DataFrame:")
Crude_suicide_df.head(30)
```

```python

merged_df = pd.merge(merged_df, Human_Resource_df, on=["Country", "Year"], how="inner")
# Display the transformed DataFrame
print("\nTransformed DataFrame:")
merged_df
```

```python
merged_df['Total_Suicide_Rates'] = merged_df['2016'] + merged_df['2015'] + merged_df['2010'] + merged_df['2000']
merged_df
```

```python
merged_df['Suicide_Rates_per_Mental_Hospitals'] = merged_df['Mental _hospitals'] / merged_df['Total_Suicide_Rates']
merged_df
```

```python
merged_df['Nurses_to_Psychiatrists_Ratio'] = merged_df['Nurses'] / merged_df['Psychiatrists']
merged_df
```

```python
merged_df['Average_Suicide_Rates'] = merged_df[['2016', '2015', '2010', '2000']].mean(axis=1)
merged_df
```

```python
merged_df['Normalized_Nurses'] = merged_df['Nurses'] / merged_df['Nurses'].max()
merged_df['Normalized_Psychiatrists'] = merged_df['Psychiatrists'] / merged_df['Psychiatrists'].max()
merged_df
```

```python
merged_df['Has_Day_Treatment_Facilities'] = merged_df['day _treatment'].apply(lambda x: 1 if x > 0 else 0)
merged_df
```

```python
age_groups = ['over_10', 'over_20', 'over_30', 'over_40', 'over_50', 'over_60', 'over_70', 'over_80']
for age_group in age_groups:
    merged_df[f'Suicide_Rates_{age_group}_Age_Group'] = merged_df[f'{age_group}']
merged_df
```

```python
threshold_high_suicide = 15  # Define a threshold for high suicide rates
merged_df['High_Suicide_Rates'] = (merged_df['Average_Suicide_Rates'] > threshold_high_suicide).astype(int)
merged_df
```

```python
merged_df['Suicide_Rates_Change'] = merged_df['2016'] - merged_df['2000']
merged_df
```

```python
merged_df['Running_Total_Suicide_Rates'] = merged_df.groupby('Country')['Total_Suicide_Rates'].cumsum()
merged_df
```

```python
selected_columns = ['Country','Sex','Total_Suicide_Rates', 'Suicide_Rates_per_Mental_Hospitals', 'Nurses_to_Psychiatrists_Ratio', 'Average_Suicide_Rates', 'Normalized_Nurses', 'Normalized_Psychiatrists','Has_Day_Treatment_Facilities']
Result_df = merged_df[selected_columns]
Result_df
```

```python
import psycopg2
from psycopg2 import sql
import csv

# Replace with your database connection details
conn = psycopg2.connect(database="Records", user="postgres", password="root", host="127.0.0.1", port="5433")
print("Database connection opened.")
cur = conn.cursor()

# Replace with your CSV file path and table name
csv_file_path = 'Result_df.csv'
table_name = 'Results'

# Open the CSV file
with open(csv_file_path, 'r') as f:
    # Check if the table exists
    cur.execute(sql.SQL("""
        SELECT EXISTS (
            SELECT 1
            FROM information_schema.tables
            WHERE table_name = {table_name}
        );
    """).format(table_name=sql.Literal(table_name)))

    table_exists = cur.fetchone()[0]

    # If the table doesn't exist, create it
    if not table_exists:
        cur.execute(sql.SQL("""
            CREATE TABLE {table_name} (col1 INT, col2 TEXT, col3 FLOAT);  -- Adjust column types as needed
        """).format(table_name=sql.Identifier(table_name)))

    # Create a temporary table
    cur.execute(sql.SQL("""
        CREATE TEMPORARY TABLE temp_table AS 
        SELECT * FROM {table_name} 
        WITH NO DATA;
    """).format(table_name=sql.Identifier(table_name)))

    # Copy the data from the CSV file to the temporary table
    cur.copy_expert(sql.SQL("""
        COPY temp_table FROM STDIN WITH CSV HEADER;
    """), f)

    # Insert the data from the temporary table into the main table
    cur.execute(sql.SQL("""
        INSERT INTO {table_name} SELECT * FROM temp_table;
    """).format(table_name=sql.Identifier(table_name)))

# Commit the transaction
conn.commit()

# Close the cursor and connection
cur.close()
conn.close()

print("Data loaded into the PostgreSQL table '{table_name}' successfully.")
```
