```python
import requests
from bs4 import BeautifulSoup

url = "http://books.toscrape.com/"
response = requests.get(url)

def extract_book_prices(response):
    soup = BeautifulSoup(response.content, "html.parser")
    prices = []

    # Find all book price elements on the page
    price_elements = soup.select('div p.price_color')

    for element in price_elements:
        # Extract the numerical value of the price
        price_text = element.text.strip().replace('Â', '')  # Clean up unwanted characters
        numerical_value = float(price_text[1:])  # Skip the currency symbol and convert to float
        prices.append(numerical_value)

    return prices

# Call the function to get book prices
book_prices = extract_book_prices(response)

# Print the result
print(book_prices)
```

```python
import requests
from bs4 import BeautifulSoup
import re

url = "http://books.toscrape.com/"
response = requests.get(url)
soup = BeautifulSoup(response.content, "html.parser")

# List to hold data values for each day
day_data_list = []

# Find all book price elements on the page
price_elements = soup.select('div p.price_color')

# Looping through each day
for price_element in price_elements:
    
    # List to hold numeric values
    numeric_values = []

    # Extract the text content of the price element
    price_text = price_element.text.strip().replace('Â', '')

    # Parse the text to get only digits
    numeric_value = '.'.join(re.findall(r"\d+", price_text))
    
    # Appending to the numeric values list
    numeric_values.append(numeric_value)

    # Appending to the data values for each day
    day_data_list.append(numeric_values)

# Print the result
print(day_data_list)
```

```python
import requests
from bs4 import BeautifulSoup

url = "http://books.toscrape.com/"
response = requests.get(url)

def scrape_books(response):
    scr = response.content
    soup = BeautifulSoup(scr, "html.parser")

    books_details = []

    book_containers = soup.find_all('article', class_='product_pod')

    for container in book_containers:
        title = container.h3.a['title']
        price = container.select('p.price_color')[0].text
        availability = container.select('p.instock.availability')[0].text.strip()
        book_url = container.h3.a['href']
        full_book_url = url + book_url

        # Extracting the book rating (example, adjust as per the actual HTML structure)
        rating = container.select('p.star-rating')[0]['class'][1]

        book_details = {
            'Title': title,
            'Price': price,
            'Availability': availability,
            'Rating': rating,
            'URL': full_book_url
        }

        books_details.append(book_details)

    return books_details

# Call the function to get the list of book details
books_data = scrape_books(response)

# Print or store the data as needed
print(books_data)
```

```python
import requests
from bs4 import BeautifulSoup
import pandas as pd

def scrape_books(response):
    scr = response.content
    soup = BeautifulSoup(scr, "html.parser")

    books_details = []

    book_containers = soup.find_all('article', class_='product_pod')

    for container in book_containers:
        title = container.h3.a['title']
        price = container.select('p.price_color')[0].text
        availability = container.select('p.instock.availability')[0].text.strip()

        # Extracting the book rating (example, adjust as per the actual HTML structure)
        rating_tag = container.select_one('p.star-rating')
        rating = rating_tag['class'][1] if rating_tag else 'N/A'

        # Extracting the book URL
        book_url = container.h3.a['href']
        full_book_url = url + book_url

        book_details = {
            'Title': title,
            'Price': price,
            'Availability': availability,
            'Rating': rating,
            'URL': full_book_url
        }

        books_details.append(book_details)

    return books_details

url = "http://books.toscrape.com/"
response = requests.get(url)

# Call the function to get the list of book details
books_data = scrape_books(response)

# Create a DataFrame
columns = ['Title', 'Price', 'Availability', 'Rating', 'URL']
Books_data = pd.DataFrame(books_data, columns=columns)

# Print or use the DataFrame as needed
print(Books_data)
```
